### Task 3

```python
x = np.linspace(-3, 3, 51)
x
```

```python
y1 = x ** 2
y2 = 2 * x + 0.5
y3 = -3 * x  - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(2, 2)

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace = 0.3, hspace = 0.3)

ax[0, 0].plot(x, y1)
ax[0, 0].set_xlim([-5, 5])

ax[0, 1].plot(x, y2)
ax[1, 0].plot(x, y3)
ax[1, 1].plot(x, y4)
    
plt.show()
```

### Task 4

```python
df = pd.read_csv('d://creditcard.csv')
plt.style.use('fivethirtyeight')
df.head()
```

```python
class_values = df['Class'].value_counts()
class_values
```

```python
class_values.plot(kind = "bar")
plt.show()
```

```python
class_values.plot(kind = "bar", logy = True)
plt.show()
```

```python
for i in range(2):
    plt.hist(df["V1"].loc[df["Class"] == i],
             bins = 20, 
             alpha = 0.5, 
             label = f"Class {i}", 
             color = "gray" if i % 2 == 0 else "red",
             density = True)
plt.xlabel("Class")
plt.legend(loc = "upper left")
plt.show()
```
